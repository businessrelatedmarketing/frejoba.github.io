```python
import pandas as pd
```

```python
df1 = pd.read_csv("./../../LCQ_1.csv")
df1["Leetcode Question Link"] = df1["Leetcode Question Link"].apply(lambda x: (x+"/").replace(" ",""))
df1
```

```python
df2  = pd.concat([df1,df], ignore_index=True)
df2
```

```python
df2.to_csv("./../../LCQ_total.csv", index=False)
```

```python
import re

# Define a function to extract the URL from the string
def extract_url(column_value):
    if column_value:
        match = re.search(r'\(([^)]+)\)', column_value)
        return match.group(1) if match else column_value
    else:
        return column_value
```

```python
df3 = pd.read_csv("../data/leetcode_repo_readme.csv")
df3["Solutions"] = df3["Solutions"].apply(lambda x: x.replace("(../","(https://github.com/fishercoder1534/Leetcode/blob/"))
df3
```

```python
df3["Leetcode Question Link"] = df3["Title"].apply(extract_url)
df3
```

```python
dff = pd.merge(df2, df3, on='Leetcode Question Link', how='left')
dff
```

```python
dff.columns
```

```python
dff = dff[['Title_x', 'Acceptance', 'Difficulty_x', 'Frequency',
       'Leetcode Question Link', 'Company_x', 'Occurrence',
       'Solutions', 'Video', 'Tag']]
dff.columns = ['Title', 'Acceptance', 'Difficulty', 'Frequency',
       'DiscussionLink', 'CompanyName', 'Occurrence',
       'Solutions', 'Video', 'Tag']
dff
```

```python
dff.to_csv("./../../LCQ_total_solInc.csv",index=False)
```

```python
def companyNameRasteurize(name):
    name = name.replace(" ", "-").replace(".","")
    return name
```

```python
dff = dff[['Title', 'Difficulty',
       'DiscussionLink', 'CompanyName',
       'Solutions', 'Video', 'Tag']]

dff["CompanyName"] = dff["CompanyName"].apply(companyNameRasteurize)
dff1 = dff.drop_duplicates()
dff1
```

```python
columns_to_check = ['CompanyName', 'DiscussionLink']

# Sort the DataFrame by other columns in descending order of precedence
dff1.sort_values(by=['Difficulty', 'Solutions'], ascending=[False, False], inplace=True)

# Drop duplicates based on the two columns and keep the first occurrence (highest precedence)
dff2 = dff1.drop_duplicates(subset=columns_to_check, keep='first')

dff2
```

```python
dff2.to_csv("./../../LCQ_total_solInc_noDuplicates.csv",index=False)
```
