```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.service import Service as FirefoxService
from selenium.webdriver.firefox.options import Options
import time


# Set the path to your geckodriver.exe
geckodriver_path = './geckodriver.exe'

# Set the path to your Firefox binary (modify this based on your system)
firefox_binary_path = 'C:/Program Files/Mozilla Firefox/firefox.exe'
    
# Configure Firefox options
firefox_options = Options()
firefox_options.binary_location = firefox_binary_path
firefox_options.add_argument('--headless')


# Create a new instance of the Firefox driver with the FirefoxService
firefox_service = FirefoxService(executable_path=geckodriver_path)

def getLeetCodeDiscussionLinks(type, query):
    # URL to crawl
    urls = ['https://leetcode.com/discuss/'+type+'?currentPage=1&orderBy=most_votes&query='+query,'https://leetcode.com/discuss/'+type+'?currentPage=1&orderBy=most_relevant&query='+query,'https://leetcode.com/discuss/'+type+'?currentPage=1&orderBy=hot&query='+query]
    driver = webdriver.Firefox(service=firefox_service, options=firefox_options)
    driver.delete_all_cookies()
    driver.maximize_window()
    link_data = []

    try:
        for url in urls:
            # Open the URL in the browser
            driver.get(url)

            # Wait for page load
            time.sleep(5)

            print(url)

            if (len(driver.find_elements(By.CLASS_NAME, "placeholder-title__3R4Q")) == 0):
                links =[x.get_attribute("href") for x  in driver.find_elements(By.CLASS_NAME,'title-link__1ay5')]
                titles =[x.text for x  in driver.find_elements(By.CLASS_NAME,'title-link__1ay5')]
                upvotes =[x.text for x  in driver.find_elements(By.CLASS_NAME,'icon-wrapper__3uKf')]

                tags = []

                for x in driver.find_elements(By.CLASS_NAME,'topic-title-wrapper__27Nt'):
                    tag = x.find_elements(By.CLASS_NAME,'tags-container__Slnd')
                    if len(tag) != 0:
                        tags.append(tag[0].text)
                    else:
                        tags.append("")
                        
                for i in range(len(links)):
                    if (query.split(" ")[0] in titles[i].lower()):
                        link_data.append({
                            "Link":links[i],
                            "Title":titles[i],
                            "Upvotes":upvotes[i*2],
                            "Views":upvotes[i*2+1],
                            "Tags":tags[i].replace("\n", "|").split("+ ")[0]
                        })

            else:
                print("No Results Page")
                return link_data

        
    except Exception as e:
        print("Issue")
        print(e)

    # Close the browser
    driver.quit()
    return link_data
```

```python
def stopDriverService():
    # Create a new instance of the WebDriver
    driver = webdriver.Firefox()
        
    # Get the underlying WebDriver service
    service = driver.service

    # Stop the WebDriver service
    service.stop()
    driver.quit()

stopDriverService()
```

```python
cp_links_dict = {}
with open("Top250Companies.txt","r") as f:
    text = f.read()

c250 = text.split("\n")
c250
```

```python
for x in c250[220:]:
    x = x.replace("-"," ")
    print(x)
    cp_links_dict[x] = {
        "SD":getLeetCodeDiscussionLinks("compensation",x + " india"),
        "IE":getLeetCodeDiscussionLinks("interview-experience",x + " india"),
        "IQ":getLeetCodeDiscussionLinks("interview-question",x + " india")
    }
```

```python
import json

file_path = 'Top250CompanyLCLinks.json'
# Step 1: Read the existing JS

try:
    with open(file_path, 'r') as file:
        # Step 2: Parse the JSON data
        existing_data = json.load(file)
except FileNotFoundError:
    # If the file doesn't exist, create an empty data structure
    existing_data = {}

# Step 3: Append new data
existing_data.update(cp_links_dict)

# Step 4: Write the updated data back to the JSON file
with open(file_path, 'w',encoding="utf-8") as file:
    json.dump(existing_data, file, indent=2)  # indent for pretty formatting, adjust as needed
```
