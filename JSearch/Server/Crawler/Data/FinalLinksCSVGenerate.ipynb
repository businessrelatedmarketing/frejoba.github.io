```python
import pandas as pd
import json
```

```python
cp_links_dict = {}
with open("Top250Companies.txt","r") as f:
    text = f.read()

c250 = text.split("\n")
c250_i = {}

for i,x in enumerate(c250):
    c250_i[x] = i

c250_i
```

```python
with open("Top250CompanyBlindLinks_final_noDuplicates.json", 'r', encoding='utf-8') as file:
    t2 = json.load(file)

with open("Top250CompanyLCLinks.json", 'r', encoding='utf-8') as file:
    t1 = json.load(file)
```

```python
def expandJsonToRows(t):
    mjson = []

    for key, values in t.items():
        mjsonk = key.replace(" ","-")

        if "SD" in values:
            for sd in values["SD"]:
                jsonObj = {
                    "Title": sd["Title"],
                    "DiscussionLink": sd["Link"],
                    "CompanyName": mjsonk,
                    "Views": sd["Views"],
                    "Type": "SD",
                }

                if "Description" in sd:
                    jsonObj["Description"] = sd["Description"]

                if "Likes" in sd:
                    jsonObj["Likes"] = sd["Likes"]

                if "Comments" in sd:
                    jsonObj["Comments"] = sd["Comments"]

                if "Upvotes" in sd:
                    jsonObj["Upvotes"] = sd["Upvotes"]

                if "Tags" in sd:
                    jsonObj["Tags"] = sd["Tags"]

                mjson.append(jsonObj)

        if "IE" in values:
            for ie in values["IE"]:
                jsonObj = {
                    "Title": ie["Title"],
                    "DiscussionLink": ie["Link"],
                    "CompanyName": mjsonk,
                    "Views": ie["Views"],
                    "Type": "IE",
                }

                if "Description" in ie:
                    jsonObj["Description"] = ie["Description"]

                if "Likes" in ie:
                    jsonObj["Likes"] = ie["Likes"]

                if "Comments" in ie:
                    jsonObj["Comments"] = ie["Comments"]

                if "Upvotes" in ie:
                    jsonObj["Upvotes"] = ie["Upvotes"]

                if "Tags" in ie:
                    jsonObj["Tags"] = ie["Tags"]

                mjson.append(jsonObj)

        if "IQ" in values:
            for iq in values["IQ"]:
                jsonObj = {
                    "Title": iq["Title"],
                    "DiscussionLink": iq["Link"],
                    "CompanyName": mjsonk,
                    "Views": iq["Views"],
                    "Upvotes": iq["Upvotes"],
                    "Tags": iq["Tags"],
                    "Type": "IQ",
                }
                mjson.append(jsonObj)

        if "C" in values:
            for c in values["C"]:
                jsonObj = {
                    "Title": c["Title"],
                    "DiscussionLink": c["Link"],
                    "CompanyName": mjsonk,
                    "Description": c["Description"],
                    "Views": c["Views"],
                    "Likes": c["Likes"],
                    "Comments": c["Comments"],
                    "Type": "C",
                }
                mjson.append(jsonObj)

    return mjson
```

```python
df = pd.DataFrame(expandJsonToRows(t1))
df.drop_duplicates(inplace=True)
df
```

```python
df.columns
```

```python
df["CompanyId"] = df["CompanyName"].apply(lambda x: c250_i[x])
df = df[['Title', 'DiscussionLink', 'Views', 'Type', 'Upvotes', 'Tags', 'CompanyId']]
df
```

```python
df.to_csv("./Links/LC_Links.csv", index=False)
```
